* dataset3 18.06~18.11 (19년 이전부터 대여소 그룹 컬럼 존재x)

```python
#dataset3 대여소그룹/ 대여소번호 페어 다시 확인
#대여소그룹 해당기간 누락된 대여소 들이 있다
#이용정보가아닌 대여소정보로 대여소그룹/ 대여소페어 추출필요
```

```python
os.chdir('../dataset3') #데이터셋마다 컬럼이 다르다 구분해서 조작 후 병합
extension = 'csv' #csv 파일 확장자 지정 
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
pair = pd.read_csv(all_filenames[0], encoding = 'cp949') #대여소 번호/ 대여소 페어 데이터프레임 추출
pair.columns = ['대여소번호', '대여소명', '자치구', '상세주소', '위도', '경도', '설치시기', '거치대수lcd', '거치대수qr', '운영방식']
#불필요 칼럼 제거 
pair = pair.drop(['대여소명', '상세주소', '위도', '경도', '설치시기', '거치대수lcd', '거치대수qr', '운영방식'], axis = 1, inplace = False)
pair = pair.dropna(subset=['대여소번호'])
pair.head()
```

```python
df1 = pd.read_csv(all_filenames[1], encoding = 'cp949')
df1.head() #대여소 그룹 필요
```

```python
#map 함수 활용 위해 사전형으로 df2를 저장
location_dict = {}
pair = pair.astype({'대여소번호' : int})

for i, i_row in pair.iterrows():
    location_dict[i_row['대여소번호']] = i_row['자치구']

df1['대여소 그룹'] = df1['대여소번호'].map(location_dict) #map함수 1차원 배열 적용 -> 사전형
```

```python
#df1 저장 및 컬럼 조정
df1 = df1.dropna(subset=['대여소 그룹'])
df1 = df1.drop(['대여소번호', '반납건수'], axis = 1)
df1 = df1[['대여소 그룹', '대여소명', '대여일자', '대여건수']]
df1.to_csv( "공공자전거 대여소별 이용정보_18.06_18.11.csv", index=False, encoding='cp949') #csv 저장
```

* dataset4 17.01~18.05

```python
os.chdir('../dataset4') #데이터셋마다 컬럼이 다르다 구분해서 조작 후 병합
extension = 'csv' #csv 파일 확장자 지정 
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
df3 = pd.read_csv(all_filenames[0], encoding = 'cp949') #17.01~17.12
df4 = pd.read_csv(all_filenames[1], encoding = 'cp949') #18.01~18.05
df3.head() #df3/df4 아래처럼 동일구성
```

```python
#컬럼 조작 및 '' 제거
df3.columns = ['대여일자', '대여소번호', '대여소명', '대여건수', '반납건수']
df4.columns = ['대여일자', '대여소번호', '대여소명', '대여건수', '반납건수']
df3['대여일자'] = df3['대여일자'].str.strip("'")
df3['대여소번호'] = df3['대여소번호'].str.strip("'")
df3['대여소명'] = df3['대여소명'].str.strip("'")
df4['대여일자'] = df4['대여일자'].str.strip("'")
df4['대여소번호'] = df4['대여소번호'].str.strip("'")
df4['대여소명'] = df4['대여소명'].str.strip("'")
df3.head()
```

```python
df3['대여소번호']=df3['대여소번호'].str.replace(pat=r'\D+', repl= r'', regex=True)
df3 = df3.drop(df3[df3['대여소번호'] == ''].index)
df3 = df3.dropna(subset=['대여소번호'])
df4['대여소번호']=df4['대여소번호'].str.replace(pat=r'\D+', repl= r'', regex=True)
df4 = df4.drop(df4[df4['대여소번호'] == ''].index)
df4 = df4.dropna(subset=['대여소번호'])
```

```python
df3 = df3.astype({'대여소번호' : int})
df3['대여소 그룹'] = df3['대여소번호'].map(location_dict)#map함수 1차원 배열 적용 -> 사전형
df3 = df3.dropna(subset=['대여소 그룹'])
df3.head()
```

```python
df4 = df4.astype({'대여소번호' : int})
df4['대여소 그룹'] = df4['대여소번호'].map(location_dict)#map함수 1차원 배열 적용 -> 사전형
df4 = df4.dropna(subset=['대여소 그룹'])
df4.head()
```

```python
#df3, df4 병합
df5 = pd.concat([df3, df4]) #3,4 병합
df5 = df5.drop(['대여소번호', '반납건수'], axis = 1)
df5 = df5[['대여소 그룹', '대여소명', '대여일자', '대여건수']]
df5.head()
```

```python
#df5 저장
df5.to_csv("공공자전거 대여소별 이용정보_17.01_18.05.csv", index=False, encoding='cp949') #병합 파일 저장
```
